```python
import pandas as pd

df = pd.read_csv("cleaned_airbnb_data.csv")
df.head()
```

```python
target = df['log_price']
features = df.drop(['price', 'log_price'], axis=1)
```

```python
from sklearn.model_selection import train_test_split

features_train, features_test, target_train, target_test = train_test_split(
    features, target, test_size=0.3, random_state=7
)
```

```python
from sklearn.ensemble import GradientBoostingRegressor

model = GradientBoostingRegressor(n_estimators=200, learning_rate=0.1, random_state=7)
model.fit(features_train, target_train)

predictions = model.predict(features_test)
```

```python
from sklearn import metrics
import numpy as np

# Since we used log_price as target, convert it back:
actual_values = np.expm1(target_test)
predicted_values = np.expm1(predictions)

print("MAE:", metrics.mean_absolute_error(actual_values, predicted_values))
print("MSE:", metrics.mean_squared_error(actual_values, predicted_values))
print("RMSE:", np.sqrt(metrics.mean_squared_error(actual_values, predicted_values)))
print("R2 Score:", metrics.r2_score(actual_values, predicted_values))
```
